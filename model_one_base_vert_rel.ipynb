# Model 1 Exploration 
I want to try to find if there is a relationship between number of vertices and the probability base for Model 1.

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("csv_files/model_one_desired_avg_deg.csv")
```

First, I am going to visualize a basic plot with no transformations.

```python
plt.scatter(x=df["num_vertices"], y=df["prob_base_num"])
plt.xlabel("# of Vertices")
plt.ylabel("Base")
plt.title("Base versus # of Vertices")
plt.savefig("imgs/model_one/base_vs_verts.png")
```

# Is it an exponential relationship?

If it is, then log(base) vs. # of vertices will have a linear relationship.

```python
plt.scatter(x=df["num_vertices"], y=np.log2(df["prob_base_num"]))
plt.xlabel("# of Vertices")
plt.ylabel("Log2(Base)")
plt.title("Log2(Base) versus # of Vertices")
plt.savefig("imgs/model_one/logbase_vs_verts.png")
```

# Is it an polynomial relationship?

If it is, then log(base) vs. log(# vertices) will have a linear relationship.

```python
plt.scatter(x=np.log2(df["num_vertices"]), y=np.log2(df["prob_base_num"]))
plt.xlabel("Log2(# of Vertices)")
plt.ylabel("Log2(Base)")
plt.title("Log2(Base) versus Log2(# of Vertices)")
plt.savefig("imgs/model_one/logbase_vs_logverts.png")
```

# Is it an logarithmic relationship?

If it is, then base vs. log(# vertices) will have a linear relationship.

```python
plt.scatter(x=np.log2(df["num_vertices"]), y=df["prob_base_num"])
plt.xlabel("Log2(# of Vertices)")
plt.ylabel("Base")
plt.title("Base versus Log2(# of Vertices)")
plt.savefig("imgs/model_one/base_vs_logverts.png")
```
